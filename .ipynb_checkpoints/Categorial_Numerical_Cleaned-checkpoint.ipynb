```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('learningSet.csv')
data.head()
```

# Separating Target, Numerical Features, and Categorical Features

```python
Y = data[['TARGET_B', 'TARGET_D']]
Y
```

```python
#Separating numericals
numerical = data.select_dtypes(np.number)
numerical = numerical.drop(columns = ['TARGET_B', 'TARGET_D'])
numerical.head()
```

# Working On Categorical Features

```python
#separating categorials
categorical = data.select_dtypes(object)
categorical.shape
```

```python
# Exclude specified variables and drop them
drop_list = ['OSOURCE', 'ZIP']
```

# 
Working on blank spaces

```python
#Replacing the blank space with A
categorical['MAILCODE'] = categorical['MAILCODE'].apply(lambda x: x.replace(" ", "A"))
```

```python
categorical['VETERANS'].value_counts()
```

```python
categorical['VETERANS'] = categorical['VETERANS'].apply(lambda x: x.replace(" ", "N"))
```

```python
# Now we can replace the rest space characters with np.NaN
categorical = categorical.apply(lambda x: x.replace(" ", np.NaN))
```

```python
categorical.head()
```

#
There are a lot of columns that have a very high percentage of null values. We can decide on a threshold and then remove those variables.n this case we will take this threshold to be 50%  

```python
categorical.isna().sum()/len(categorical)
```

```python
nulls_percent_df = pd.DataFrame(categorical.isna().sum()/len(categorical)).reset_index()
nulls_percent_df
nulls_percent_df.columns = ['column_name', 'nulls_percentage']
nulls_percent_df
```

```python
nulls_percent_df[nulls_percent_df['nulls_percentage']!=0]
```

# 
columns that have over 50% missing values.

```python
columns_above_threshold = nulls_percent_df[nulls_percent_df['nulls_percentage']>0.5]
columns_above_threshold['column_name']
```

```python
drop_columns_list = list(columns_above_threshold['column_name'])
print(drop_columns_list)
```

```python
# Add elements from drop_list to drop_columns_list
drop_columns_list.extend(drop_list)
```

#
Dropping the columns

```python
categorical = categorical.drop(drop_columns_list,axis=1)
categorical.shape
```

# 
Cleaning the other columns

```python
# group all state < 2500 as 'other
df = pd.DataFrame(categorical['STATE'].value_counts()).reset_index()

df.columns = ['state', 'count']
other_states = list(df[df['count']<2500]['state'])
#other_states
def clean_state(x):
    if x in other_states:
        return 'other'
    else:
        return x
    
categorical['STATE'] = list(map(clean_state, categorical['STATE']))
```

```python
# Replace null values in the GENDER column
print("Before replacing null values in GENDER:")
print(categorical['GENDER'].value_counts())
categorical['GENDER'] = categorical['GENDER'].fillna('F')
```

```python
# Reduce the number of categories in the GENDER column
categorical['GENDER'] = categorical['GENDER'].apply(lambda x: 'other' if x not in ['M', 'F'] else x)
```

```python
# Check the reduced categories
print("After reducing categories in GENDER:")
print(categorical['GENDER'].value_counts())
```

```python
categorical
```

```python
# Check for null values across all columns
null_values_check = categorical.isnull()

# Count the number of null values in each column
null_values_count = null_values_check.sum()

# Print the result
print("Number of null values in each column:")
print(null_values_count)
```

```python
categorical['HOMEOWNR'].value_counts(dropna=False)
```

```python
# Replacing NAN with U..Nan also means 'we don't know'
categorical['HOMEOWNR'] = categorical['HOMEOWNR'].fillna('U') 
```

```python
categorical['DOMAIN'].value_counts(dropna=False)
```

```python
drop_list2=[]
```

```python
#The percentage of null values for each column
num_nulls=pd.DataFrame(numerical.isna().sum()*100/len(numerical)).reset_index()
num_nulls.columns = ['column_name', 'nulls_percentage']

#there is a lot more columns than for the categoricals so I'll set a threshold at 25% (to keep less columns)
columns_above_threshold = num_nulls[num_nulls['nulls_percentage']>25]
len(columns_above_threshold['column_name'])
```

```python
drop_list = list(columns_above_threshold['column_name'])
drop_list

# I am removing WEALTH from the droplist
drop_list.remove('WEALTH1')
drop_list.remove('WEALTH2')

#droping the rest
numerical.drop(drop_list, axis=1, inplace= True)
numerical.shape
```

```python
df = pd.DataFrame(numerical.isna().sum()).reset_index()
df.columns = ['column_name', 'nulls']
df[df['nulls']>0]
```

# ADATE
Dropping ADATE

```python
numerical['ADATE_7'].value_counts()
```

```python
drop_list = []
for col_name in numerical.columns:
    if "ADATE" in col_name:
        drop_list.append(col_name)   
numerical.drop(drop_list, axis=1, inplace=True)
```

# Handling AGE,WEALTH1 ,WEALTH2,INCOME,CLUSTER2

```python
#For WEALTH, AGE, INCOME and CLUSTER2, I am replacing the NAN with the mean

replace_mean_list=['WEALTH1', 'WEALTH2', 'AGE', 'INCOME', 'CLUSTER2']
for items in replace_mean_list:
    numerical[items] = numerical[items].fillna(round(np.mean(numerical[items]),2))
```

# Checking MSA,DMA,ADI
observing the above column values, deciding to replace NaN with the mode

```python
numerical['MSA'].value_counts()
```

```python
numerical['ADI'].value_counts()
```

```python
numerical['DMA'].value_counts()
```

```python
#replacing NaN with the mode 
numerical['MSA'] = numerical['MSA'].fillna(0.0)
numerical['ADI'] = numerical['ADI'].fillna(13.0)
numerical['DMA'] = numerical['DMA'].fillna(803.0)
```

# Handling TIMELAG and NEXTDATE

```python
# replacing Nan with 0
numerical['TIMELAG'] = numerical['TIMELAG'].fillna(0.0)
```

```python
numerical['NEXTDATE'] = numerical['NEXTDATE'].fillna(0.0)
```

# 
Rechecking if there are Nan values remaining

```python
df = pd.DataFrame(numerical.isna().sum()).reset_index()
df.columns = ['column_name', 'nulls']
df[df['nulls']>0]
```
